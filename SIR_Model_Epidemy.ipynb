On souhaite appliquer quelques méthodes numériques vues en cours pour simuler la propagation
au cours du temps d’une épidémie (le coronavirus par exemple) au sein d’une population
donnée P. On utilise le modèle, appelé SIR, qui suppose que la population est divisée en 3 compartiments :

— S : les individus Sains ou Susceptibles d’être infectés.

— I : ceux qui sont Infectés.

— R : ceux qui sont Rétablis.

----
Ces populations varient bien sûr dans le temps. Notons les noterons respectivement S(t), I(t),R(t).
Le système d’équations différentielles auxquelles obéissent ces populations s’écrit:

$$ \left\{
\begin{array}{ccc}
{\large{\frac{ds}{dt}}} & = & - \beta IS \\
 \large{\frac{dI}{dt}} & = &  \beta IS  -(\frac{1}{\lambda}+\mu)I \\
 \large{\frac{dR}{dt}} & = & \large{\frac{I}{\lambda}}  \\
 \end{array}
\right.

$$

où $\beta$  > 0 désigne le taux d’incidence, $\lambda$ est le nombre de jours en moyenne durant lesquels un
individu reste malade et $\mu$ est un paramètre dont la valeur est proportionnelle à la virulence de
l’agent contaminant.
Bien sûr, au système ci-dessus, il faut adjoindre des conditions initiales S(0) = S0, I(0) = I0,
R(0) = R0. On suppose ici que R0 = 0, qui exprime le fait qu’aucun individu n’est immunisé
au début de la période d’infection.

$\bf\underline{Euler ~~Explicite} :$

$$ \left\{
\begin{array}{ccc}

 R_{n+1}=R_{n} + \frac{1}{\lambda}I_{n}\Delta t

 \end{array}
\right.

$$

 $$ \left\{
\begin{array}{ccc}

S_{n+1} & =  &  S_{n} & - & \beta SI \Delta T \\
I_{n+1} & = & I_{n} & + & (\beta S I-(\large{\frac{I}{\lambda}}- \mu)I) \Delta T\\

 \end{array}
\right.

$$

 $$ \left\{
\begin{array}{ccc}

S & =  &  S_{n} / (1+\beta I_{n}\Delta T) \\
I & = & (1+(\large{\frac{I}{\lambda}}+ \mu -\beta S_{n}) \Delta T\\
\end{array}
\right.

$$
Obtenus par résolution par point fixe.

# **Implementation**

```
import numpy as np 
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```
# fonction qui permet de tracer les coubres du modele SIR
def graphe(s,i,r,p,t):
  plt.plot(t , s,'r',label='Non infected')
  plt.plot(t , i,'g',label='Infected')
  plt.plot(t , r,'k',label='Recovery')
  plt.plot(t, p ,'b',label='Population')
  plt.legend()
  plt.grid()
  plt.show()
```

## **Euler Explicite**

```
@interact(s0=(0.0,1.0,0.01),lam=(5,30),beta=(0.0,1.0,0.01),mu=(0.0,0.1,0.01))
def SIR_model_explicite(s0,lam=5,beta=0.5,mu=0.00):
  T = 30
  t = np.arange(0,T)
  n=len(t)
  delta_t = T / 30
  s = np.zeros(n)
  i = np.zeros(n)
  r = np.zeros(n)
  s[0] ,i[0] , r[0]= s0, 1-s0, 0 # r0 = 0 et i0 = 1 - s0
  for j in range(1,n):
    s[j] = s[j-1] - beta * s[j-1] * i[j-1] * delta_t
    i[j] = i[j-1] + ( beta * s[j-1] * i[j-1] - ((1/lam) + mu ) * i[j-1]  ) * delta_t
    r[j] = r[j-1] + (1/lam) * i[j-1]  * delta_t
  p = s+i+r
  graphe(s , i , r , p , t)
```

## **Euler Implicite**

```
@interact(s0=(0.0,1.0,0.01),lam=(5,30),beta=(0.0,1.0,0.01),mu=(0.0,0.1,0.01))
def SIR_model_implicite(s0=0.7,lam=5,beta=0.5,mu=0.0):
  T = 30
  t = np.arange(0,T)
  n=len(t)
  delta_t = T / 30
  s = np.zeros(n)
  i = np.zeros(n)
  r = np.zeros(n)
  s[0] ,i[0] , r[0]= s0, 1-s0, 0 # r0 = 0 et i0 = 1 - s0
  for j in range(1,n):
    # point fixe 
    s_n = s[j-1] / ( 1 + beta * i[j-1] * delta_t )
    i_n = i[j-1] / ( 1 + ( 1/lam + mu - beta * s[j-1] ) * delta_t )

    s[j] = s[j-1] - beta * s_n * i_n * delta_t
    i[j] = i[j-1] + ( beta * s_n * i_n - ((1/lam) + mu ) * i_n  ) * delta_t
    r[j] = r[j-1] + (1/lam) * i_n * delta_t
  p = s + i + r
  graphe(s , i , r , p , t)
  
```

## **Runge Kutta d'ordre 2 ( Heun )**

```
@interact(s0=(0.0,1.0,0.01),lam=(5,30),beta=(0.0,1.0,0.01),mu=(0.0,0.1,0.01))
def SIR_model_heun(s0=0.7,lam=5,beta=0.5,mu=0.0):
  T = 30
  t = np.arange(0,T)
  n=len(t)
  delta_t = T / 30
  s = np.zeros(n)
  i = np.zeros(n)
  r = np.zeros(n)
  s[0] ,i[0] , r[0]= s0, 1-s0, 0 # r0 = 0 et i0 = 1 - s0
  for j in range(1,n):
    s_n = s[j-1] - beta * s[j-1] * i[j-1] * delta_t
    i_n = i[j-1] + ( beta * s[j-1] * i[j-1] - (1/lam) * i[j-1] ) * delta_t

    s[j] = s[j-1] - (delta_t / 2) * (beta * ( s[j-1] * i[j-1] + s_n * i_n ))
    i[j] = i[j-1] + (delta_t/2) * ( beta * (s_n * i_n + s[j-1] * i[j-1] ) - ((1/lam) + mu ) * ( i_n + i[j-1] ))
    r[j] = r[j-1] + (delta_t / 2 ) * (1/lam ) * ( i_n + i[j-1] )
  p = s + i + r
  graphe(s , i , r , p , t)
```

## **Runge Kutta d'ordre 4**

```
@interact(s0=(0.0,1.0,0.01),lam=(5,30),beta=(0.0,1.0,0.01),mu=(0.0,0.1,0.01))
def SIR_model_kutta_4(s0=0.7,lam=5,beta=0.5,mu=0.0):
  T = 30
  t = np.arange(0,T)
  n=len(t)
  delta_t = T / 30
  s = np.zeros(n)
  i = np.zeros(n)
  r = np.zeros(n)
  s[0] ,i[0] , r[0]= s0, 1-s0 , 0 # r0 = 0 et i0 = 1 - s0
  for j in range(1,n):

    k1_sn = delta_t * (- beta * s[j-1] * i[j-1] )
    k1_in = delta_t * ( beta * s[j-1] * i[j-1] - (1/lam + mu ) * i[j-1] )
    k1_rn = delta_t * (1 / lam ) * i[j-1]
                       
    s_n = s[j-1] + k1_sn / 2
    i_n = i[j-1] + k1_in / 2
    r_n = r[j-1] + k1_rn / 2

    k2_sn = delta_t * ( - beta * s_n * i_n )
    k2_in = delta_t * ( beta * s_n * i_n - (1/lam + mu) * i_n )
    k2_rn = delta_t * (1/lam) * i_n


    s_n = s_n + k2_sn / 2
    i_n = i_n + k2_in / 2
    r_n = r_n + k2_rn / 2

    k3_sn = delta_t * ( - beta * s_n * i_n )
    k3_in = delta_t * ( beta * s_n * i_n - (1/lam + mu ) * i_n )
    k3_rn = delta_t * (1/lam) * i_n

    s_n = s_n + k3_sn
    i_n = i_n + k3_in 
    r_n = r_n + k3_rn

    k4_sn = delta_t * ( - beta * s_n * i_n )
    k4_in = delta_t * ( beta * s_n * i_n - (1/lam + mu) * i_n )
    k4_rn = delta_t * (1/lam) * i_n


    s[j] = s[j-1] + (k1_sn + k4_sn) / 6 + ( k2_sn + k3_sn )/3
    i[j] = i[j-1] + (k1_in + k4_in) / 6 + ( k2_in + k3_in )/3
    r[j] = r[j-1] + (k1_rn + k4_rn) / 6 + ( k2_rn + k3_rn )/3
  p = s + i + r
  graphe(s , i , r , p , t)
```
